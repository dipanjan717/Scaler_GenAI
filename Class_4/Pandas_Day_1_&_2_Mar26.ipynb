```python
samples_orders = pd.DataFrame({
    'product_id': ['P001', 'P002', 'P003'],
    'user_id': ['U001', 'U002', 'U003'],
    'review_id': ['R001', 'R002', 'R003']
})

samples_products = pd.DataFrame({
    'product_id': ['P001', 'P002', 'P004'],
    'product_name': ['Product A', 'Product B', 'Product D'],
    'category': ['Category 1', 'Category 2', 'Category 3']
})
```

```python
final = pd.merge(orders, products[["x", "y", "z"]], on='product_id', how='left')
```

```python
final.head()
```

```python
final.head()
```

```python
def extract_price(price_str):
        return float(str(price_str).replace('₹', '').replace(',', ''))
```

```python
final['actual_price_fixed'] = final['actual_price'].apply(extract_price)
```

```python
final.head()
```

```python
def splitter(price_str):
        return str(price_str).split('₹')
```

```python
final['checker'] = final['discounted_price'].apply(splitter)
```
