```python
def categorize_rating(rating):
    "Categorize ratings into high or low."""
    if rating >= 4.5:
        return "High"
    else:
        return "Low"
```

```python
ratings
```

```python
categorize_rating(ratings)
```

```python
categorize_rating_np = np.vectorize(categorize_rating)
```

```python
categorize_rating_np(ratings)
```
