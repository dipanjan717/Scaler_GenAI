```python
monthly_ratings= df.groupby ('month_year')['rating_filled'].mean()
```

```python
monthly_ratings= monthly_ratings.reset_index()
```

```python
sns.lineplot(x="month_year", y="rating_filled",
             data=monthly_ratings)
plt.xticks(rotation=90)

```

```python
monthly_ratings['month_year']= pd.to_datetime(monthly_ratings['month_year'], format= "%B %Y")
```

```python
monthly_ratings.sort_values('month_year', inplace=True)
```

```python
monthly_ratings
```

```python
sns.lineplot(x="month_year", y="rating_filled",
             data=monthly_ratings)
plt.xticks(rotation=90)
plt.show()
```
